###   *Question # 07*
***Write a program which takes 5 inputs from user for different subject’s marks, total it
and generate mark sheet using grades ?***
---

```python
Mathematics = int(input("Mathematics marks : "))
English =int(input("English marks : "))
Islamiat =int(input("Islamiat marks : "))
Physics =int(input("Physics marks : "))
Chemistry  =int(input("Chemistry marks : "))
totalMarks = Mathematics + English + Islamiat + Physics + Chemistry 
print("\nTotal Marks : ",totalMarks)

Percentage = int(totalMarks/500 * 100 )
print("Percentage : ",Percentage,"%")

Grade = "" ;

if Percentage >= 85:
    Grade = " A1 \t Excellent !"
elif Percentage >=75 and Percentage <= 85:
    Grade = " A \t Good !"
elif Percentage >=65 and Percentage <= 75:
    Grade = " B \t Average"
elif Percentage >=55 and Percentage <= 65:
    Grade = " C \t Below Average"  
elif Percentage >=45 and Percentage <= 55:
    Grade = " D \t Needs Improvement !"
else :
    Grade : " F, \t FAIL!"
    
        
print("Your Grade is : ",Grade)

    

```

###   *Question # 10*
***Write a Python program to sum all the numeric items in a list?***
___

```python
def sum(number) :
    sum = 0
    for y in number :
        sum+= y
    return sum

PyList = [25,50,20,5]
print(sum(PyList))
```
